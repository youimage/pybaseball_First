```python
from pybaseball import statcast_pitcher

# 投手の選手IDと期間を指定してデータを取得
data = statcast_pitcher('2023-03-30', '2023-10-01', 660271)

# 合計失点を計算
total_runs_allowed = data['events'].isin(['single', 'double', 'triple', 'home_run']).sum()

# 合計被安打数を計算
total_hits_allowed = data['events'].isin(['single', 'double', 'triple', 'home_run']).sum()

# 合計四死球数を計算
total_walks_allowed = data['events'].isin(['walk', 'walked', 'intent_walk', 'hit_by_pitch']).sum()

# 合計奪三振数を計算
total_strikeouts = data['events'].isin(['strikeout']).sum()

# 合計被安打数を計算
total_hits_allowed = data['events'].isin(['single', 'double', 'triple', 'home_run']).sum()

print("合計失点:", total_runs_allowed)
print("合計被安打数:", total_hits_allowed)
print("合計四死球数:", total_walks_allowed)
print("合計奪三振数:", total_strikeouts)
print("合計被安打数:", total_hits_allowed)
```

```python
#whip=(与四球＋被安打）÷投球回
whip = (total_walks_allowed + total_hits_allowed)/132

#k/9＝奪三振×9÷投球回
k_nine =  total_strikeouts * 9 /132

#B/9=与四死球×9÷投球回
b_nine = total_walks_allowed *9 /132

print("whip:", whip)
print("K/9):", k_nine)
print("B/9:", b_nine)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from pybaseball import statcast_pitcher

# 大谷投手のデータを取得
data = statcast_pitcher('2023-03-30', '2023-10-01', 660271)

# ファールボールを含めた球種別のストライク率を計算
pitch_type_total = data['pitch_type'].value_counts()
pitch_type_strikes = data[data['pitcher'] == 660271].groupby('pitch_type')['description'].apply(lambda x: (x == 'foul') | (x == 'called_strike')).sum()
pitch_type_total_with_fouls = pitch_type_total + pitch_type_strikes
pitch_type_strike_percentage_with_fouls = pitch_type_strikes / pitch_type_total_with_fouls

print("球種別ストライク率:\n", pitch_type_strike_percentage_with_fouls)

# 結果を円グラフで視覚化
plt.figure(figsize=(8, 8))
plt.pie(pitch_type_strike_percentage_with_fouls, labels=pitch_type_strike_percentage_with_fouls.index, autopct='%1.1f%%')
plt.title('球種別ストライク率（ファールボール含む）')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from pybaseball import statcast_pitcher

# 藤波投手のデータを取得
data = statcast_pitcher('2023-03-30', '2023-10-01', 660271)

# ストレートの投球データを抽出
fastball_data = data[data['pitch_type'] == 'FF']

# ストレートのリリースポイントを取得
release_points = fastball_data[['release_pos_x', 'release_pos_y', 'release_pos_z']].values

# ストライクとボールのデータを抽出
strike_data = fastball_data[fastball_data['description'] == 'called_strike']
ball_data = fastball_data[fastball_data['description'] == 'ball']

# ストライクとボールのリリースポイントを取得
strike_release_points = strike_data[['release_pos_x', 'release_pos_y', 'release_pos_z']].values
ball_release_points = ball_data[['release_pos_x', 'release_pos_y', 'release_pos_z']].values

# プロット
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# ストライクのリリースポイントを赤色でプロット
ax.scatter(strike_release_points[:, 0], strike_release_points[:, 1], strike_release_points[:, 2], c='red', label='Strike')

# ボールのリリースポイントを青色でプロット
ax.scatter(ball_release_points[:, 0], ball_release_points[:, 1], ball_release_points[:, 2], c='blue', label='Ball')

# x軸、y軸、z軸のラベルを設定
ax.set_xlabel('Release Position X')
ax.set_ylabel('Release Position Y')
ax.set_zlabel('Release Position Z')

# グラフのタイトルを設定
plt.title('Release Points of Fastballs (Strikes vs Balls)')

# グリッドを表示
ax.grid(True)

plt.legend()
plt.tight_layout()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pybaseball import statcast_pitcher

# 藤波投手のデータを取得
data = statcast_pitcher('2023-03-30', '2023-10-01', 660271)

# ストレートの投球データを抽出
fastball_data = data[data['pitch_type'] == 'FF']

# 各試合ごとのストレートの平均球速とストライク率を計算
game_dates = fastball_data['game_date'].unique()
avg_velocity_per_game = []
strike_rate_per_game = []

for date in game_dates:
    game_data = fastball_data[fastball_data['game_date'] == date]
    avg_velocity = game_data['release_speed'].mean()
    strike_count = (game_data['description'] == 'called_strike').sum()
    total_pitches = len(game_data)
    strike_rate = strike_count / total_pitches
    avg_velocity_per_game.append(avg_velocity)
    strike_rate_per_game.append(strike_rate)

# ピアソンの相関係数を計算
correlation_coefficient = np.corrcoef(avg_velocity_per_game, strike_rate_per_game)[0, 1]

# グラフで表示
plt.figure(figsize=(10, 6))
plt.scatter(avg_velocity_per_game, strike_rate_per_game, marker='o')
plt.title('Correlation between Fastball Velocity and Strike Rate')
plt.xlabel('Average Velocity (mph)')
plt.ylabel('Strike Rate')
plt.grid(True)
plt.tight_layout()

# 相関係数を表示
plt.text(90, 0.4, f'Correlation Coefficient: {correlation_coefficient:.2f}', fontsize=12)

plt.show()
```

```python
import numpy as np

# ストライクの各座標軸の平均値と標準偏差を計算
strike_mean_x = np.mean(strike_release_points[:, 0])
strike_mean_y = np.mean(strike_release_points[:, 1])
strike_mean_z = np.mean(strike_release_points[:, 2])
strike_std_x = np.std(strike_release_points[:, 0])
strike_std_y = np.std(strike_release_points[:, 1])
strike_std_z = np.std(strike_release_points[:, 2])

# ボールの各座標軸の平均値と標準偏差を計算
ball_mean_x = np.mean(ball_release_points[:, 0])
ball_mean_y = np.mean(ball_release_points[:, 1])
ball_mean_z = np.mean(ball_release_points[:, 2])
ball_std_x = np.std(ball_release_points[:, 0])
ball_std_y = np.std(ball_release_points[:, 1])
ball_std_z = np.std(ball_release_points[:, 2])

print("Strike Release Point Mean (X):", strike_mean_x)
print("Strike Release Point Mean (Y):", strike_mean_y)
print("Strike Release Point Mean (Z):", strike_mean_z)
print("Strike Release Point Standard Deviation (X):", strike_std_x)
print("Strike Release Point Standard Deviation (Y):", strike_std_y)
print("Strike Release Point Standard Deviation (Z):", strike_std_z)

print("Ball Release Point Mean (X):", ball_mean_x)
print("Ball Release Point Mean (Y):", ball_mean_y)
print("Ball Release Point Mean (Z):", ball_mean_z)
print("Ball Release Point Standard Deviation (X):", ball_std_x)
print("Ball Release Point Standard Deviation (Y):", ball_std_y)
print("Ball Release Point Standard Deviation (Z):", ball_std_z)
```

```python
# ストライクの各座標軸のばらつき具合を数値化
strike_variability_x = strike_std_x / strike_mean_x
strike_variability_y = strike_std_y / strike_mean_y
strike_variability_z = strike_std_z / strike_mean_z

# ボールの各座標軸のばらつき具合を数値化
ball_variability_x = ball_std_x / ball_mean_x
ball_variability_y = ball_std_y / ball_mean_y
ball_variability_z = ball_std_z / ball_mean_z

print("Strike Release Point Variability (X):", strike_variability_x)
print("Strike Release Point Variability (Y):", strike_variability_y)
print("Strike Release Point Variability (Z):", strike_variability_z)

print("Ball Release Point Variability (X):", ball_variability_x)
print("Ball Release Point Variability (Y):", ball_variability_y)
print("Ball Release Point Variability (Z):", ball_variability_z)
```
